## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Import Dataset

```python
df = pd.read_csv("drug200.csv")
df
```

```python
df.shape
```

## Data Preprocessing

```python
df.Sex.unique()
```

```python
df.BP.unique()
```

```python
df.Cholesterol.unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder().fit(df.Sex.unique())
df.Sex = le_sex.transform(df.Sex)
```

```python
from sklearn.preprocessing import LabelEncoder
le_bp = LabelEncoder().fit(df.BP.unique())
df.BP = le_bp.transform(df.BP)
```

```python
from sklearn.preprocessing import LabelEncoder
le_choles = LabelEncoder().fit(df.Cholesterol.unique())
df.Cholesterol = le_choles.transform(df.Cholesterol)
```

```python
df
```

## Slipt train and Test

```python
X = df.iloc[:, : -1].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 3)
```

```python
X_train.shape
```

## Fit the Model

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(criterion = "entropy", max_depth = 4)
dtree
```

```python
dtree.fit(X_train, y_train)
```

## Predict by Model

```python
y_pred = dtree.predict(X_test)
```

```python
print(y_pred[:5])
print(y_test[:5])
```

## Evaluation

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

## Visualization

```python
from  io import StringIO
import pydotplus
import matplotlib.image as mpimg
from sklearn import tree
%matplotlib inline 
```

```python
dot_data = StringIO()
filename = "drugtree.png"
featureNames = df.columns[0:5]
out=tree.export_graphviz(dtree, feature_names = featureNames, out_file = dot_data, class_names = np.unique(y_train), filled=True,  special_characters=True,rotate=False)  
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png(filename)
img = mpimg.imread(filename)
plt.figure(figsize=(100, 200))
plt.imshow(img,interpolation='nearest')
```
